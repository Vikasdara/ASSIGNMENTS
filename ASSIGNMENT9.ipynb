```python
from sklearn.cluster import KMeans
```

```python
X = df[['Alcohol', 'Malic']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model=KMeans(n_clusters=3)
df["cluster"]=model.fit_predict(X_scaled)
df
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(X_scaled,df['cluster'])
print(score)
```

```python
if 'Type' in df.columns:
    X = df.drop(['Type', 'cluster', 'cluster_pca'], axis=1, errors='ignore')
else:
    X = df.drop(['cluster', 'cluster_pca'], axis=1, errors='ignore')
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
model=KMeans(n_clusters=3,random_state=32)
df["cluster_pca"]=model.fit_predict(X_pca)
df
```

```python
score=silhouette_score(X_pca,df['cluster_pca'])
print(score)
```

```python
scores=[]
for i in range(2,11):
  model=KMeans(n_clusters=i)
  df["cluster"]=model.fit_predict(X_scaled)
  sc=silhouette_score(X_scaled,df['cluster'])
  scores.append(sc)

```

```python
scores
```

```python
plt.figure(figsize=(7,5))
plt.plot(range(2,11), scores, 'o-', linewidth=2)
plt.title('Silhouette Score vs Number of Clusters (after PCA)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
#1.IN ORIGINAL DATASET THE SCORE IS 48,BUT AFTER APPLYING PCA THE SCORE IS 33
#2.Major clusters (high-density regions) often remain consistent between both datasets.
#3.Removes noise and correlations â†’ simpler, faster clustering.
#4.SPEED(CLUSTERING ON ORIGINAL (SLOWER)),(Clustering after PCA(FAST))
```

```python
# Dataset has many correlated numeric features,Apply PCA first, then cluster,PCA reduces redundancy and noise.
```
