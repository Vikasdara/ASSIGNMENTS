```python
categorical_col=df.select_dtypes(include="object").columns
```

```python
for col in categorical_col:
  print(f"Column: {col}")
  print(df[col].value_counts())
  print("\n")
```

```python
unique_counts = {col: df[col].nunique() for col in categorical_col}
```

```python
unique_counts
```

```python
one_hot_cols=[col for col,count in unique_counts.items() if count<5]
```

```python
one_hot_cols
```

```python
label_encoded=[col for col,count in unique_counts.items() if count>=5]
```

```python
label_encoded
```

```python
hot_encoded=pd.get_dummies(df,columns=one_hot_cols,drop_first=True)
hot_encoded['capital_diff'] = df['capital_diff']
```

```python
hot_encoded
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
label_encoders = {}
for col in label_encoded:
    le = LabelEncoder()
    hot_encoded[col] = le.fit_transform(hot_encoded[col])
    label_encoders[col] = le
```

```python
label_encoders
```

```python
hot_encoded.head()
```

```python
def categorize(hours):
  if hours<35:
    return"part-time"
  elif hours <=40:
    return "full-time"
  else:
    return "over-time"
```

```python
categorize(34)
```

```python
df["works-hour-category"]=df["hours_per_week"].apply(categorize)
```

```python
df['capital_diff'] = df['capital_gain'] - df['capital_loss']
```

```python
from sklearn.ensemble import IsolationForest
```

```python
islo_features = hot_encoded[[
    'age',
    'education_num',
    'hours_per_week',
    'capital_diff',
    'capital_loss',
    'workclass',
    'occupation',
    'relationship',
    'sex_ Male'
]]
```

```python
clf=IsolationForest(random_state=10,contamination=.01)
clf.fit(islo_features)
```

```python
y_predict_outliers=clf.predict(islo_features)
y_predict_outliers
```

```python
df['scores']=clf.decision_function(islo_features)
df['anamoly']=clf.predict(islo_features)
```
