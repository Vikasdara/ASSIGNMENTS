```python
outliers={}
for col in df.select_dtypes(include=np.number).columns:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  outliers[col]=df[(df[col]<lower_bound) | (df[col]>upper_bound)].index.tolist()
  print(f"{col}:{len(outliers[col])}")
```

```python
df_capped=df.copy()
for col in df_capped.select_dtypes(include=np.number).columns:
  Q1=df_capped[col].quantile(0.25)
  Q3=df_capped[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  df_capped[col]=np.where(df_capped[col]<lower_bound,lower_bound,np.where(df_capped[col]>upper_bound,upper_bound,df_capped[col]))

```

```python
x=df.iloc[:, 1:16]
```

```python
x
```

```python
y=df['type']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_train_predict=model.predict(x_train)
y_test_predict=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
print("Training_accuracy_score",np.round(accuracy_score(y_train,y_train_predict),2))
print("Testing_accuracy_score",np.round(accuracy_score(y_test,y_test_predict),2))
print("Training_precision_score",np.round(precision_score(y_train,y_train_predict, average='weighted'),2))
print("Testing_precision_score",np.round(precision_score(y_test,y_test_predict, average='weighted'),2))
print("Training_recall_score",np.round(recall_score(y_train,y_train_predict, average='weighted'),2))
print("Testing_recall_score",np.round(recall_score(y_test,y_test_predict, average='weighted'),2))
print("Training_f1_score",np.round(f1_score(y_train,y_train_predict, average='weighted'),2))
print("Testing_f1_score",np.round(f1_score(y_test,y_test_predict, average='weighted'),2))
```

```python
from sklearn.model_selection import ShuffleSplit,cross_validate
s=ShuffleSplit(n_splits=10,test_size=0.2,random_state=42)

```

```python
cv_result=cross_validate(model,x,y,cv=s,scoring='accuracy', return_train_score=True)
```

```python
train_score=np.round(cv_result['train_score'],2)
test_score=np.round(cv_result['test_score'],2)
```

```python
train_score
```

```python
test_score
```

```python
k_train_score=[]
k_test_score=[]
for i in range(5,19,2):
  model=KNeighborsClassifier(n_neighbors=i)
  cv_strategy = ShuffleSplit(n_splits=10,test_size=0.2,random_state=42)
  cv_result=cross_validate(model,x,y,cv=s,scoring='accuracy', return_train_score=True)
  train_score=cv_result['train_score']
  test_score=cv_result['test_score']
  k_train_score.append(np.round(train_score, 2))
  k_test_score.append(np.round(test_score, 2))
```

```python
k_train_score
```

```python
plt.figure(figsize=(8,10))

k_values = list(range(5,19,2))
mean_k_train_score = [np.mean(scores) for scores in k_train_score]

plt.scatter(k_values, mean_k_train_score)
plt.plot(k_values, mean_k_train_score, color='red')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Mean Training Accuracy Score')
plt.title('Mean Training Accuracy Score vs. Number of Neighbors (k)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,10))

k_values = list(range(5,19,2))
mean_k_test_score = [np.mean(scores) for scores in k_test_score]

plt.scatter(k_values, mean_k_test_score)
plt.plot(k_values, mean_k_test_score, color='red')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Mean Test Accuracy Score')
plt.title('Mean Test Accuracy Score vs. Number of Neighbors (k)')
plt.grid(True)
plt.show()
```

```python
print('Analysis of Mean Test Accuracy Score vs. Number of Neighbors (k):')
print('The plot above illustrates the mean test accuracy score for various values of k (number of neighbors).')
print('As observed, the test accuracy generally decreases as k increases. This suggests that a smaller k leads to better generalization performance on this dataset.')
print('Finding the optimal k value typically involves identifying the point where test accuracy is maximized. In this case, it appears to be at a lower k value.')
```
